```python
#Encode for categorical value

from sklearn.preprocessing import OneHotEncoder as prep  #(Import libriay for box plot)

encd = prep(sparse=False)
encd_Column = encd.fit_transform(dataSet[['Region','Trunk Calling Facility','Voice Messaging']])

#Creating data frame from above column
encd_df = pd.DataFrame(encd_Column, columns = encd.get_feature_names_out(['Region','Trunk Calling Facility','Voice Messaging']))

cleaned_dataSet = pd.concat([dataSet.drop(['Region','Trunk Calling Facility','Voice Messaging'], axis=1), encd_df], axis=1)

cleaned_dataSet.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, f1_score
import numpy as np
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
```

```python
#creating a copy of data 
cleaned_dataSet_copy = cleaned_dataSet.copy()
```

```python
# applying label encoding to categorical values
# labelEncoder                                = LabelEncoder()
# cleaned_dataSet['Trunk Calling Facility']   = labelEncoder.fit_transform(cleaned_dataSet['Trunk Calling Facility'])
# cleaned_dataSet['Voice Messaging']          = labelEncoder.fit_transform(cleaned_dataSet['Voice Messaging'])
# dataSet
```

```python
# one hot encoding for region - as region consists of more than one type of values
dataSet = pd.get_dummies(dataSet, columns=['Region'], drop_first=True)
# dataSet

# Columns to be scaled
columsnWithLargeNumericalValues = [ 'Tenure', 'Neighborhood', 'Number voice messages', 'Minutes Peak Hrs', 
                                   'Calls Peak Hrs', 'Bill Peak Hrs', 'Minutes Off Peak', 'Calls Off Peak', 
                                   'Bill Off Peak', 'Minutes Night', 'Calls Night', 'Bill Night', 
                                   'Trunk Call Minutes', 'Trunk Calls', 'Trunk Call Bill',
                                   'Contact for Grievances/Changes']
```

### Trial 1 - Using Initialize StandardScaler

```python
# Trial 1 - Using Initialize StandardScaler
scaler = StandardScaler()
# using StandardScaler for bring the large numerical values close to mean & SD
cleaned_dataSet[columsnWithLargeNumericalValues] = scaler.fit_transform(cleaned_dataSet[columsnWithLargeNumericalValues])
dataSet
```

Here we see that some of the normalised values are negatives are some are more than 1

```python
features    = cleaned_dataSet.drop(['Acct Closed?'],axis=1)
target      = cleaned_dataSet[['Acct Closed?']]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=60)
```

```python
model           = LogisticRegression()
model.fit(X_train, y_train)
y_predicted     = model.predict(X_test)
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_predicted):.2f}")
print(f"Precision: {precision_score(y_test, y_predicted):.2f}")
print(f"Recall: {recall_score(y_test, y_predicted):.2f}")
print(f"F1 Score: {f1_score(y_test, y_predicted):.2f}")
print(f"AUC: {roc_auc_score(y_test, y_predicted):.2f}")
```

```python
cm = confusion_matrix(y_test, y_predicted)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d')
```

### Trial 2 - Using Initialize MinMaxScaler


```python
# Trial 2 - Using Initialize MinMaxScaler
scaler  = MinMaxScaler() 
# using MinMaxScaler for bring the large numerical values between 0 & 1
cleaned_dataSet[columsnWithLargeNumericalValues] = scaler.fit_transform(cleaned_dataSet[columsnWithLargeNumericalValues])
cleaned_dataSet
```
